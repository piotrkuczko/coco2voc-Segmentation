```python
from coco2voc import *
from PIL import Image
import numpy as np
import cv2
```

```python
class_changer  = {
    1: 15, #person
    2: 2, #bicycle
    3: 7, #car
    4: 14, #motorcycle, motorbike
    5: 1, #airplane, aeroplane
    6: 6, #bus
    7: 19, #train
    9: 4, #boat
    16: 3, #bird
    17: 8, #cat
    18: 12, #dog
    19: 13, #horse
    20: 17, #sheep
    21: 10, #cow
    44: 5, #bottle
    62: 9, #chair
    64: 16, #potted plant
    65: 18, #sofa-bed
    67: 11, #dinning table
    72: 20, #tv
}
palette = [[0,0,0],
[128,0,0], 
[0,128,0],
[128,128,0],
[0,0,128],
[128,0,128], 
[0,128,128],
[128,128,128],
[64,0,0],
[192,0,0],
[64,128,0],
[192,128,0],
[64,0,128],
[192,0,128],
[64,128,128],
[192,128,128],
[0,64,0],
[128,64,0],
[0,192,0],
[128,192,0],
[0,64,128]] + [[255, 255, 255]] * 235
```

```python
palette = [value for color in palette for value in color]
```

```python
# !!Change paths to your local machine!!
annotations_file = './MSCOCO2017/annotations/instances_train2017.json'
labels_target_folder = './output'
data_folder = './MSCOCO2017/train2017'


# Convert n=25 annotations
coco2vocSegmentation(annotations_file, labels_target_folder, n=25, class_changer=class_changer, palette=palette)
```
